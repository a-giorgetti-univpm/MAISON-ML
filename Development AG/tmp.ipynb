```python
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
pd.DataFrame([], columns=["Macro-F1", "Balanced Accuracy", "Macro Recall", "Macro Precision"])
```

```python
df = pd.DataFrame({
    "text": [
        "machine learning",
        "deep learning",
        "machine intelligence",
        "learning systems"
    ]
})
```

```python
count_vec = CountVectorizer()

count_matrix = count_vec.fit_transform(df["text"])
count_df = pd.DataFrame(
    count_matrix.toarray(),
    columns=[f"bow_{w}" for w in count_vec.get_feature_names_out()]
)
```
